```python
df['Ranking'].hist()

def outliers_z_score(ys, threshold = 3):
    mean_y = np.mean(ys)
    std_y = np.std(ys)
    z_scores = [(y - mean_y) / std_y for y in ys]
    return np.where(np.abs(z_scores) > threshold)[0]

len(outliers_z_score(df['Ranking']))
```

```python
def outliers_iqr(ys):
    quartile_1, quartile_3 = np.percentile(ys, [25, 75])
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5)
    upper_bound = quartile_3 + (iqr * 1.5)
    print('les', len(np.where(ys < lower_bound)[0]))
    print('gr', len(np.where(ys > upper_bound)[0]))
    return np.where((ys > upper_bound) | (ys < lower_bound))[0]
a = outliers_iqr(df['Ranking'])
```

```python
from scipy.stats import norm

def estimate_gaussian(dataset):
    mu = np.mean(dataset, axis=0)
    sigma = np.cov(dataset.T)
    return mu, sigma
    
def get_gaussian(mu, sigma):
    distribution = norm(mu, sigma)
    return distribution

def get_probs(distribution, dataset):
    return distribution.pdf(dataset)

est = estimate_gaussian(df['Ranking'])
dist = get_gaussian(est[0], est[1])
probs = get_probs(dist, df['Ranking'])
len(np.where(probs < 0.0000001)[0])
```

```python
est = estimate_gaussian(df['Rating'])
dist = get_gaussian(est[0], est[1])
probs = get_probs(dist, df['Rating'])
len(np.where(probs < 0.001)[0])
```

```python
est = estimate_gaussian(df['Number of Reviews'])
dist = get_gaussian(est[0], est[1])
probs = get_probs(dist, df['Number of Reviews'])
len(np.where(probs < 0.00001)[0])
```

```python
def price_identification(value):
    if value == np.nan:
        return 1
    elif value == '$':
        return 2
    elif value == '$$ - $$$':
        return 3
    elif value == '$$$$':
        return 4
df['Price Range'] = df['Price Range'].apply(lambda x: price_identification(x))
df['Price Range'] = df['Price Range'].fillna(1)
```

# Разбиваем датафрейм на части, необходимые для обучения и тестирования модели

```python
# Х - данные с информацией о ресторанах, у - целевая переменная (рейтинги ресторанов)
X = df.drop(['Cuisine Style', 'Reviews', 'Rating'], axis = 1)
y = df['Rating']
```

```python
# Загружаем специальный инструмент для разбивки:
from sklearn.model_selection import train_test_split
```

```python
# Наборы данных с меткой "train" будут использоваться для обучения модели, "test" - для тестирования.
# Для тестирования мы будем использовать 25% от исходного датасета.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# Создаём, обучаем и тестируем модель

```python
# Импортируем необходимые библиотеки:
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
# Создаём модель
regr = RandomForestRegressor(n_estimators=100)

# Обучаем модель на тестовом наборе данных
regr.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = regr.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```
